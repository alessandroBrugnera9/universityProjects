# 1. Introdução

A escolha de um intervalo de amostragem $h$ é crucial para o funcionamento do 
sistema de controle.

O intervalo de amostragem deve ser suficientemente pequeno para perceber todas
as frequências envolvidas tanto da planta a ser controlada quanto dos sinais
de distúrbio e ruídos de medida. Na prática, entretanto, um intervalo de amostragem muito
pequeno pode sobrecarregar o sistema computacional.

Deve ser lembrado que entre os intervalos de amostragem o sistema funciona em malha aberta.

As equações de diferenças no domínio do tempo discreto são algoritmos de integração numérica.
O limite de frequência estabelecido pelo Teorema de Nyquist não é suficiente para
garantir uma integração numérica precisa. Usualmente é necessário uma taxa
de amostragem $f_a=1/h$ várias vezes maior que a maior frequência contida nos sinais do sistema.

A transformação de pólos do domínio $s$ para o domínio $z$ depende do intervalo de amostragem $h$,
dessa forma, variações de $h$ podem tornar um sistema mais ou menos estável.

A escolha do período de amostragem pode ser feita com base numa variável adimensional e que possui uma interpretação física. Para sistemas oscilatórios é natural a normalização com base no período de oscilação, para sistemas não oscilatórios, o tempo de subida do sistema é o fator de normalização adequado.

Define-se o número de períodos de amostragem por tempo de subida, $N_r$:
\begin{equation}
N_r = \frac{t_r}{h}
\end{equation}
onde $t_r$ é o tempo de subida. Para sistemas de $1^a$ ordem, o tempo de subida pode ser substituído pela constante de tempo do sistema. Portanto, é razoável escolher $N_r$ entre 4 e 10 para o caso de projeto do controle ter sido realizado em tempo discreto. Contudo, se o projeto do controlador foi realizado diretamente em tempo contínuo, $N_r$ deve ficar entre 10 e 20.

Para um sistema de segunda ordem com coeficiente de amortecimento $\zeta$ e frequência natural $\omega_n$, o tempo de subida é dado por:
\begin{equation}
t_r = {\omega_n}^{-1} e ^{\phi/\tan{\phi}}
\end{equation}
onde $\zeta = \tan{\phi}$.  Para um coeficiente de amortecimento em torno de $\zeta=0.7$, isto resulta em:
\begin{equation}
\omega_n h\simeq 0.1\;ou\; 0.05
\end{equation}
Onde $\omega_n$ é dado em $\text{rad}/s$.

Uma questão importante a ser considerada é que os polos e zeros no domínio $z$ dependem do período do amostragem. Como se sabe, os polos do domínio $s$ são mapeados no domínio $z$ através da seguinte relação:
\begin{equation}
z = e ^{sh} \label{eqtransp} \;\;(1)
\end{equation}

Na medida em que o período de amostragem $h \rightarrow 0$, ou seja, diminui, os poólos estáveis em $s$ (pólos com parte real negativa) tendem rapidamente para $z=1$. Paradoxalmente pode-se esperar que conforme o período de amostragem diminui a aproximação do sistema contínuo se torna mais precisa, no entanto períodos de amostragem pequenos podem introduzir distorções significativas no comportamento dinâmico do sistema. Isto decorre principalmente em razão do número finito de algarismos utilizado em um computador digital. Assim, por exemplo, considere um pólo em tempo discreto localizado em $z = 0,999991$, o que é uma posição muito próxima de $z =1$ que é o limite da estabilidade, após inúmeros cálculos realizados pelo controlador, a cada período de amostragem, podem ocorrer erros de arredondamento que levam o sistema facilmente para a instabilidade.

Uma outra forma de escolher o período de amostragem é utilizar a Equação (1) de modo que os pólos do sistema de malha aberta em tempo discreto fiquem em torno de 0.95. Assim,
invertendo a Equação (1), tem-se:
\begin{equation}
h=\frac{\ln (0.95)}{p}
\end{equation}
onde $p$ é o pólo mais rápido da malha aberta. Esse método garante que os pólos do sistema em
tempo discreto tenham um número de dígitos significativos suficiente de forma que não sejam
arredondados para a unidade devido a erros de quantização.

Alternativamente, existem proposições para escolha do intervalo de amostragem baseado na largura de banda $\omega_B$ de malha fechada, por exemplo:
$$
6 \leq \frac{\omega_s}{\omega_B} \leq 40,
$$
onde $\omega_s$ é a frequência angular de amostragem, i.e., $\omega_s=2\pi/h$.



# 2. Controle de velocidade de um motor CC com controlador PI - Influência do intervalo de amostragem $h$

Nessa seção objetiva-se analisar a influência do intervalo de amostragem $h$ no desempenho do sistema de controle em malha fechada.

Para isso utilizaremos como exemplo o sistema de controle de velocidade de um motor CC utilizando um controlador PI. Um exemplo semelhante foi visto na Experiência 4.

O projeto do controlador PI é fornecido a priori:
$$
H(s) = K_p(1+\frac{1}{T_i})
$$

O controlador no domínio $z$ é obtido discretizando-se o controlador através do método Backward-Diffference:
$$
H(z) = K_p\left( 1 + \frac{h}{T_i}\frac{z}{z-1}\right)
$$

Da Experiência 3, sabemos que a função de transferência da velocidade angular de um motor elétrico CC pode ser reduzido a seguinte expressão:
$$
\frac{\Omega(s)}{U(s)} = G_\omega(s) = \frac{K_\omega}{T s+1}
$$
A função de transferência discreta do sistema considerando o segurador de Ordem zero na entrada do sistema pode ser calculado como:
$$
G_\omega (z) = (1-z^{-1}) {\cal{L}} \left\{ \frac{G_\omega(s)}{s} \right\},
$$

Deseja-se observar a variação do desempenho do sistema de controle de tempo discreto variando-se o valor do intervalo de amostragem $h$ sempre tendo o sistema de controle no domínio do tempo contínuo como referência.

```python
import numpy as np
import control as co_general
import matplotlib.pyplot as plt
import control.matlab as co
plt.close('all')
# Escolher essa opcao para o plot em janela fora do notebook
#%matplotlib qt
# Opcao para plot interativo dentro do notebook (mais dificil controlar o zoom)
#%matplotlib notebook 
# plot inline
%matplotlib inline
#
# Definicao do intervalo de amostragem h
#
# h=0.01, 0.02, 0.05, 0.1
#
h = 0.01
#
# Velocidade do motor CC
# Gw = Kw/(Ts+1)
#
s = co.tf('s')
print('--------------------------------')
print('planta do sistema no dominio s')
print('--------------------------------')
Kw = 1.0
T  = 0.25
Gw = Kw/(T*s+1)
print('Gw(s) = ',Gw)
print('--------------------------------')
print('planta do sistema no dominio z')
print('discretizacao com zoh')
print('--------------------------------')
Gwz = co.c2d(Gw,h,'zoh')
print('Gwz(z) = ',Gwz)
#
# Controlador PI 
#                     1
#   H(s) = Kp*( 1 + ----- )    
#                    Ti*s
Kp = 3.0
Ti = 0.1
H = Kp*(1+1/(Ti*s))
print('--------------------------------')
print('Controlador PI')
print('--------------------------------')
print('H(s) = ',H)
print('H(s) Polos = ',co.pole(H))
print('H(s) Zeros = ',co.zero(H))
# malha aberta
GwH = Gw*H
print('\nCaracteristicas da malha aberta---------->\n')
print('GwH =',GwH)
# malha fechada
cloop = co.feedback(GwH,1)
print('\nCaracteristicas da malha fechada---------->\n')
print('cloop =',cloop)
print('cloop Polos =',co.pole(cloop))
print('cloop Zeros =',co.zero(cloop))
#
# Definicao do vetor de tempo com espacamento
# igual ao intervalo de amostragem h
#
t = [x*h for x in range(0,200)]
#
# Resposta a degrau unitario
#
y,t=co.step(cloop,t)
co.stepinfo(cloop)
print('----------------------------------------------\n')
print('Controlador PI utilizando Backward-Difference')
print('\n----------------------------------------------')
# Parte Integral
HBackDiff_I = co.tf([h, 0],[Ti,-Ti],h)
# somatoria das partes P + I
HBackDiff   = Kp*(1 + HBackDiff_I)
#
# LR de tempo discreto
#
#C =[j*0.01 for j in range(0,800)]
#co.rlocus(Gwz*HBackDiff,grid=False,xlim=[-3.2,+3.2])
#
# Definicao da malha fechada
#
cloopHBackDiff = co.feedback(Gwz*HBackDiff,1)
# Malha fechada
print('\nCaracteristicas da malha fechada---------->\n')
print('cloopHBackDiff(z) =',cloopHBackDiff)
print('cloopHBackDiff(z) Polos =',co.pole(cloopHBackDiff))
print('cloopHBackDiff(z) Zeros =',co.zero(cloopHBackDiff))
co.pzmap(cloopHBackDiff,grid=True)
# Calculo da resposta a degrau unitario
ybd,t = co.step(cloopHBackDiff,t)
plt.figure(2)
# dominio do tempo continuo
plt.plot(t,y)
# dominio do tempo discreto
plt.step(t,ybd)
plt.grid()
```

# 3. Algoritmos de discretização: robustez em relação à variação do intervalo de amostragem $h$

O método mais intuitivo para o projeto de sistemas de controle é através do método indireto aonde o projeto é realizado no domínio $s$ e depois realiza-se a discretização do controlador transformando-o para o domínio $z$.
Sabemos que vários métodos de discretização podem ser utilizados. Todos os métodos de discretização podem ser interpretados como a transformação da equação diferencial do controlador no domínio do tempo contínuo em uma equação de diferenças no domínio do tempo discreto que realiza a integração numérica da equação diferencial com o intervalo de amostragem $h$ podendo ser interpretado como o passo de integração.

Os seguintes métodos serão aqui considerados:

- Método de Euler (Forward-Difference):
$$
s = \frac{z-1}{h}
$$

- Método Backward-Difference
$$
s = \frac{z-1}{zh}
$$

- Método de Tustin (Transformação Bilinear)
$$
s = \frac{2}{h}\frac{z-1}{z+1}
$$

Uma questão importante a ser considerada é que esses métodos mapeiam o semi-plano esquerdo (região estável)
do plano $s$ em regiões distintas do plano $z$ como observado na figura abaixo.

```python
import numpy as np
import control as co_general
import matplotlib.pyplot as plt
import control.matlab as co
plt.close('all')
# Escolher essa opcao para o plot em janela fora do notebook
%matplotlib qt
# Opcao para plot interativo dentro do notebook (mais dificil controlar o zoom)
#%matplotlib notebook 
# opcao inline
#%matplotlib inline
#
# Definicao do intervalo de amostragem h
h = 0.05
#
# Planta do sistema no dominio s
#
s = co.tf('s')
print('--------------------------------')
print('planta do sistema no dominio s')
print('--------------------------------')
G = 1/(s+1)**3
print('G(s) = ',G)
print('--------------------------------')
print('planta do sistema no dominio z')
print('discretizacao com zoh')
print('--------------------------------')
Gz = co.c2d(G,h,'zoh')
print('Gz(z) = ',Gz)

#
# Controlador PID H(s) sem filtro 
#                     1        
#   H(s) = Kp*( 1 + ----- + Td*s )    
#                    Ti*s     
# 
Kp = 2.0
Ti = 1.0
Td = 1.0
# Controlador PID
H = Kp*(1+1/(Ti*s)+Td*s)
print('--------------------------------')
print('Controlador PID dominio s')
print('--------------------------------')
print('H(s) = ',H)
#print('H(s) Polos = ',co.pole(H))
print('H(s) Zeros = ',co.zero(H))
# mala aberta
GH = G*H
print('\nCaracteristicas da malha aberta dominio s ---------->\n')
print('\nGH =',GH)
print('\nGH Polos =',co.pole(GH))
print('\nGH Zeros =',co.zero(GH))
# malha fechada
cloop = co.feedback(GH,1)
print('\nCaracteristicas da malha fechada dominio s---------->\n')
print('\ncloop =',cloop)
print('\ncloop Polos =',co.pole(cloop))
print('\ncloop Zeros =',co.zero(cloop))
#
# Definicao do vetor de tempo com intervalo
# igual ao intervalo de amostragem h
# tmax = tempo de simulacao maximo
tmax = 50
nmax = int(tmax/h)      # numero de intervalos de amostragem
t = [x*h for x in range(0,nmax)]
#
# Resposta a degrau unitarios
y,t = co.step(cloop,t)
# Nao ha plot aqui so no final
#
# Conversao da planta para o dominio z
# considerando ZOH
#
Gz = co.c2d(G,h,'zoh')
print('Planta discretizada com ZOH Gz = ',Gz)

#
# Controladores discretizados com metodos:
# Euler, Backward-Difference, Tustin
#
# Retirar os comentarios aonde for adequado
#
# Metodo de Euler - Forward Difference
#
#print('----------------------------------------------\n')
#print('Controlador PID utilizando Euler/Forward-Difference')
#print('\n----------------------------------------------')
# Parte integral _I
Hd_I = co.tf([h],[Ti,-Ti],h)
# Parte derivativa sem filtro _D
Hd_D = co.tf([Td,-Td],[h],h)
# somatoria das partes P + I + D
Hd = Kp*(1 + Hd_I + Hd_D)

#
# Controlador Backward Difference
#
#print('----------------------------------------------\n')
#print('Controlador PID utilizando Backward-Difference')
#print('\n----------------------------------------------')
#Hd_I = co.tf([h, 0],[Ti,-Ti],h)
# Parte derivativa sem filtro _D
#Hd_D = co.tf([Td,-Td],[h, 0],h)
# somatoria das partes P + I + D
#Hd   = Kp*(1 + Hd_I + Hd_D)

#
# Aproximacao de Tustin
#
#print('----------------------------------------------\n')
#print('Controlador PID utilizando Metodo de Aprox. de Tustin')
#print('\n----------------------------------------------')
#Hd_I = co.tf([h, h],[2*Ti,-2*Ti],h)
# Parte derivativa sem filtro _D
#Hd_D = co.tf([2*Td,-2*Td],[h, h],h)
# somatoria das partes P + I + D
#Hd  = Kp*(1 + Hd_I + Hd_D)

print('\nControlador PID caracteristicas ----------->\n')
print('Hd(z) = ',Hd)

# Para o metodo de Euler a funcao que estima os polos nao
# funciona talvez pq tem mais zeros do que polos ?
#print('Hd(z) Polos = ',co.pole(Hd))
print('Hd(z) Zeros = ',co.zero(Hd))
# Malha Aberta Gz*Hd
print('\nCaracteristicas da malha aberta ---------->\n')
print('Gz*Hd(z) =',Gz*Hd)
print('\nGz*Hd(z) Polos =',co.pole(Gz*Hd))
print('\nGz*Hd(z) Zeros =',co.zero(Gz*Hd))
cloopHd= co.feedback(Gz*Hd,1)
# Malha fechada
print('\nCaracteristicas da malha fechada---------->\n')
print('cloopHd(z) =',cloopHd)
print('\ncloopHd(z) Polos =',co.pole(cloopHd))
print('\ncloopHd(z) Zeros =',co.zero(cloopHd))

# Mapa de polos e zeros no dominio z
co.pzmap(cloopHd,grid=True)

# Resposta a degrau unitario
yd,t = co.step(cloopHd,t)
print('\nFIM \n----------------------------------------------\n')

# Plots no dominio do tempo
plt.figure(3)
plt.plot(t,y,label='continuo')
plt.step(t,yd,label='discreto')
plt.legend()
plt.grid()
```
